```python
import pandas as pd
import plyvel

df = pd.read_csv('/Volumes/Transcend/work/mecab-ipadic-neologd/seed/all.csv',
                 header=None, usecols=[0, 11])

db = plyvel.DB('/Volumes/Transcend/work/mecab-ipadic-neologd/seed/yomi.ldb', create_if_missing=True)
with db.write_batch() as wb:
    for i, s in df.iterrows():
        wb.put(str(s[0]).encode('utf-8'), str(s[11]).encode('utf-8'))

db.close()
```
